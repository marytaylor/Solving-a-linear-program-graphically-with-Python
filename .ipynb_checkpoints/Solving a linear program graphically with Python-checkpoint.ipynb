```python
import matplotlib.pyplot
from matplotlib.pyplot import *
import numpy
from numpy import arange
```

```python
figure()
x = arange(-10, 10, 1)
y = arange(-10, 10, 1)

# Max: 5x1+4x2
y1 = (-5*x)/4
# constraint: x1+2x2<= 6
y2 = (6-x)/2
# constraint: -2x1+x2<=4
y3 = 4 + 2*x
# constraint: 5x1+3x2<=15
y4 = (15 - 5*x)/3
# constraint: x1,x2 >=0

xlim(-1, 8)
ylim(-1, 8)
hlines(0, -10, 10, color='k')
vlines(0, -10, 10, color='k')
grid(True)

xlabel('x1-axis')
ylabel('x2-axis')

plot(x, y1, color='b')
plot(x, y2, color='r')
plot(x, y3, color='m')
plot(x, y4, color='y')

title('Objective function 5x1 + 4x2 with the following contraints')
legend(['5x1+4x2', 'x1+2x2<=6', '-2x1+x2<=4', '5x1+3x2<=15'])

x = [0.0, 12/7, 3.0, 0.0]
y = [3.0, 15/7, 0.0, 0.0]
fill(x,y,'b')
show()
```
